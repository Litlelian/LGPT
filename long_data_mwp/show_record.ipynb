```python
import pickle
import matplotlib.pyplot as plt

# 指定.pkl檔案的路徑
file_path = 'record/fake_german_tree_md3_float3_tree_r3_f3_float3.pkl'

# 使用二進制模式打開檔案
with open(file_path, 'rb') as file:
    # 使用pickle模組的load函式讀取檔案內容
    backup_list_mwp = pickle.load(file)
```

```python
%matplotlib inline
print(f'epoch: {len(backup_list_mwp["train"]["loss"])}')

plt.figure(figsize=(10,6))
plt.title('mwp loss')
plt.plot(range(len(backup_list_mwp['train']['loss'])), backup_list_mwp['train']['loss'], label='train')
plt.plot(range(len(backup_list_mwp['dev']['loss'])), backup_list_mwp['dev']['loss'], label='dev')
plt.legend()
plt.show()

plt.figure(figsize=(10,6))
plt.title('mwp acc')
plt.plot(range(len(backup_list_mwp['train']['acc'])), backup_list_mwp['train']['acc'], label='train')
plt.plot(range(len(backup_list_mwp['dev']['acc'])), backup_list_mwp['dev']['acc'], label='dev')
plt.legend()
plt.show()
```
